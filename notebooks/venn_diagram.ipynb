# ベン図

```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn2, venn3

plt.rcParams["font.size"] = 14
```

```python
set1 = set(["A", "B", "B", "C", "C"])
set2 = set(["A", "A", "B", "D", "E"])
set3 = set(["A", "B", "E", "F", "G"])
```

## 2つの集合
- `subsets`に`set`を設定する方が楽
- ラベルと色を設定できる

```python
venn2(
    subsets=[set1, set2],
    set_labels=("set1", "set2"),
    set_colors=("r", "g")
)
plt.show()
```

## 3つの集合

```python
venn3(
    subsets=[set1, set2, set3],
    set_labels=("set1", "set2", "set3"),
    set_colors=("r", "g", "b")
)
plt.show()
```
