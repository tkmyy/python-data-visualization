```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(
    style="whitegrid",
    font_scale=1.2
)

# サンプルデータ
df = pd.read_csv("data/train.csv")
df.head()
```

## jointplot
- **散布図**と**各軸のデータの分布**を同時に描画
-　引数
    - `hue`: グルーピング
    - `joint_kws`: 散布図のパラメータ
        - `s`: マーカーサイズ
    - `marginal_kws`: ヒストグラムのパラメータ
        - `bins`: ビン数
    - `palette`: マーカーの色を指定

```python
sns.jointplot(
    data=df,
    x="Age",
    y="Fare",
    joint_kws={"s": 20, "color": "dodgerblue"},
    marginal_kws={"bins": 30, "color": "dodgerblue"}
)
plt.show()
```

```python
sns.jointplot(
    data=df,
    x="Age",
    y="Fare",
    hue="Survived",
    palette={0: "dodgerblue", 1: "tomato"},
    joint_kws={"s": 20}
)
plt.show()
```

```python
# 凡例を移動
p = sns.jointplot(
    data=df,
    x="Age",
    y="Fare",
    hue="Survived",
    palette={0: "dodgerblue", 1: "tomato"},
    joint_kws={"s": 20}
)
sns.move_legend(p.ax_joint, "upper left")

# タイトル・軸ラベル
p.fig.suptitle("title")
p.ax_joint.set_xlabel("x label")
p.ax_joint.set_ylabel("y label")
p.fig.tight_layout()

plt.show()
```
