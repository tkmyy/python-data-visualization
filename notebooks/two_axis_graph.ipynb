# 2軸グラフ

```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams["font.size"] = 14
```

```python
idx = ["A", "B", "C", "D"]
val = np.array([40, 30, 20, 10])
ratio = np.cumsum(val)/sum(val)
```

## グリッド無し

```python
fig, ax = plt.subplots(1, 1)

ax.bar(idx, val, color="deepskyblue")

ax2 = ax.twinx()
ax2.plot(idx, ratio, c="tomato", marker="o")

plt.show()
```

## グリッド有り
- 左右の目盛り位置がずれるため工夫が必要
- 目盛り幅 (`set_sticks`) と範囲 (`set_ylim`) を指定することで対処

```python
plt.rcParams["axes.axisbelow"] = True
```

```python
fig, ax = plt.subplots(1, 1)

ax.bar(idx, val, color="deepskyblue")
ax.set_yticks([i*10 for i in range(5)])
ax.set_ylim(0, 40+10*1/2) # 幅×1/2で余白を持たせる

ax2 = ax.twinx()
ax2.plot(idx, ratio, c="tomato", marker="o")
ax2.set_yticks([0.4+i*0.15 for i in range(5)])
ax2.set_ylim(0.4, 1.0+0.15/2)
ax2.grid(axis="y")

plt.show()
```
